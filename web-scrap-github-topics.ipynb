# Top Repositories for GitHub Topics

## Outline:
    
- we're going to  scrape https://github.com/topics
- we'll get a list  of topics. For each topic,we will get topic title,topic page url and topic description
- For each topic ,we will get the top 25 repositories from topic page

# We can use BeautifulSoup  library to parse and extract information

```python
#!pip install beautifulsoup4 --upgrade --quite
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(page_contents,'lxml')
#soup=BeautifulSoup(page_contents,'html.parser')
```

```python
print(soup.prettify())
```

```python
topics=soup.find_all('div',class_='py-4')
```

```python
len(topics)
```

```python
topics[0].find('p' ,class_='f3').text.strip()
```

```python
topics[0].find('p' ,class_='f5').text.strip()
```

```python
"https://github.com" + topics[0].find('a' ,class_='no-underline flex-1 d-flex flex-column')['href']
```

```python
topic_title=[]
topic_url=[]
topic_desc=[]

for i in topics:
    topic_title.append(i.find('p' ,class_='f3').text.strip())
    topic_desc.append(i.find('p' ,class_='f5').text.strip())
    topic_url.append( "https://github.com" + 
                     i.find('a' ,class_='no-underline flex-1 d-flex flex-column')['href'] )
    


    
#print(topic_title)
#print(topic_url)
#print(topic_desc)
```

```python
#creating dataframe from lists we obtained

topic_df=pd.DataFrame({'topic_title':topic_title,
                   'topic_url':topic_url,
                   'topic_decription':topic_desc})
```

## finally doing it for each topic

```python
final=pd.DataFrame()

for j in topic_url:
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    response_topic=requests.get(j ,headers=headers)
    
    soup_topic=BeautifulSoup(response_topic.text,'lxml')
    
    h3_tags=soup_topic.find_all('h3',class_='f3 color-fg-muted text-normal lh-condensed')
    
    span_tag=soup_topic.find_all('span' ,class_="Counter js-social-count")
     
    user_name=[]
    repo_name=[]
    repo_url=[]
    stars=[]
    
    for m in h3_tags:
        user_name.append(m.find_all('a')[0].text.strip())
        repo_name.append(m.find_all('a')[1].text.strip())
        repo_url.append("https://github.com" + 
                            m.find_all('a')[1]['href'])
    
    for n in span_tag:
        stars.append(n.text)
        
    repo=pd.DataFrame({"user_name":user_name,
                    "repo_name":repo_name,
                    "stars":stars,
                    "repo_url":repo_url,
                    "topic_url":j
                        })
    
    final=final.append(repo,ignore_index=True)

    
final
```
